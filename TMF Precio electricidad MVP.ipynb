Let's start using some basic models, like the LinearRegression

```python
X = prices2013_brent_weather.drop(["Total €/MWh","Day", "Altura"], axis = 1)
y = prices2013_brent_weather["Total €/MWh"]
```

```python
from sklearn import model_selection
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, shuffle = False)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(y_pred - y_test, bins=20);
```

```python
lr.score(X_test, y_test)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, y_pred)
```

```python
#MSE value
metrics.mean_squared_error(y_test, y_pred)
```

```python
#RMSE value
metrics.mean_squared_error(y_test, y_pred, squared=False)
```

It looks like the Linear Regression model works very well when we only have a few months for 2013. We'll see if this is the same when we have all years.
Also, I think this is because we're using all the variables that make up the final electricity price in the model, let's try removing some of those.

```python
X = prices2013_brent_weather.drop(["Total €/MWh","Day", "Altura", "Mercado diario €/MWh",
                                  "Mercado intradiario €/MWh", "Restricciones €/MWh",
                                  "Procesos OS €/MWh", "Garantía potencia Pagos capacidad €/MWh"], axis = 1)
y = prices2013_brent_weather["Total €/MWh"]

from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, shuffle = False)
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(y_pred - y_test, bins=20);
```

```python
lr.score(X_test, y_test)
```

As expected, the predictions are much worse using a linear regression if we don't use all the components that make up the final total price.

```python
from sklearn import metrics

metrics.mean_absolute_error(y_test, y_pred)
```

```python
#MSE value
metrics.mean_squared_error(y_test, y_pred)
```

```python
#RMSE value
metrics.mean_squared_error(y_test, y_pred, squared=False)
```
